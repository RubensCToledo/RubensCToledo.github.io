```python
STQ = []
STQ_sn = []
y2bar = np.mean(df['media'].values)
sn2bar = np.mean(df['sn'].values)

for i in range(len(df)):
    STQ.append((ym[i] - y2bar)**2)
    STQ_sn.append((sn[i] - sn2bar)**2)
STQ = np.sum(STQ)
STQ = np.mean(STQ)
STQ_sn = np.sum(STQ_sn)
STQ_sn = np.mean(STQ_sn)
```
